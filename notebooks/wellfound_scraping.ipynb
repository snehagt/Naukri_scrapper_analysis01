```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import pandas as pd
import time

chrome_options = Options()
chrome_options.add_argument("--headless")  # Run in headless mode (no GUI)
driver = webdriver.Chrome(options=chrome_options)

base_url = "https://wellfound.com/role/software-engineer?page="

job_titles = []
company_names = []
salary_ranges = []
skills_list = []

for page_num in range(1, 3):  # There are 367 pages
    url = base_url + str(page_num)
    driver.get(url)
    
    time.sleep(5)
    
    job_elements = driver.find_elements(By.CSS_SELECTOR, "a.styles_component__UCLp3.styles_defaultLink__eZMqw.styles_jobTitle___jT4l")
    
    for job_element in job_elements:
        job_title = job_element.text
        print(job_title)
        
        parent_element = job_element.find_element(By.XPATH, '..')  # Go up to the parent element
        print(parent_element)
        
        try:
            company_name = parent_element.find_element(By.CSS_SELECTOR, "h4.styles_name__rSxBl").text
        except:
            company_name = "N/A"
        
        try:
            salary_range = parent_element.find_element(By.CSS_SELECTOR, "span.styles_locationAndCompensation__vDfgW").text
        except:
            salary_range = "N/A"
        
        job_link = job_element.get_attribute('href')
        driver.get(job_link)
        
        time.sleep(3)
        
        try:
            skills_elements = driver.find_elements(By.CSS_SELECTOR, "div.mb-1.styles-module_component__2E93_.inline-flex.flex-row.items-center.mr-2.last:mr-0.rounded-full.align-middle.bg-gray-200.text-gray-700.gap-2.text-xs.px-3.py-1")
            skills = [skill.text for skill in skills_elements]
        except:
            skills = []
        
        job_titles.append(job_title)
        company_names.append(company_name)
        salary_ranges.append(salary_range)
        skills_list.append(skills)
        
        driver.back()
        
        time.sleep(3)
    
    print(f"Processed page {page_num} of 367")

df = pd.DataFrame({
    'Job Title': job_titles,
    'Company Name': company_names,
    'Salary Range': salary_ranges,
    'Skills': skills_list
})

df.to_csv('wellfound_jobs_with_skills.csv', index=False)

driver.quit()

print(df)
```
